```python
# Re-import after reset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson

# Parameter
λ = 8  # rata-rata kedatangan per jam
jam_per_hari = 12
hari = 30
total_jam = jam_per_hari * hari

# 1. Simulasi jumlah pelanggan per jam selama 30 hari
np.random.seed(42)
simulasi_kedatangan = np.random.poisson(lam=λ, size=total_jam)

# 2. Visualisasi histogram frekuensi kedatangan
plt.figure(figsize=(10, 6))
sns.histplot(simulasi_kedatangan, bins=range(0, max(simulasi_kedatangan)+2), kde=False, color='skyblue', stat='density', label='Simulasi')
x_vals = np.arange(0, max(simulasi_kedatangan)+1)
plt.plot(x_vals, poisson.pmf(x_vals, λ), marker='o', color='red', label='Distribusi Poisson Teoritis')
plt.title("Distribusi Jumlah Kedatangan Pelanggan per Jam")
plt.xlabel("Jumlah Pelanggan per Jam")
plt.ylabel("Frekuensi (Normalisasi)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Hitung probabilitas
prob_0 = np.mean(simulasi_kedatangan == 0)
prob_gt_10 = np.mean(simulasi_kedatangan > 10)

# Probabilitas Teoritis
prob_0_teoritis = poisson.pmf(0, λ)
prob_gt_10_teoritis = 1 - poisson.cdf(10, λ)

prob_0, prob_gt_10, prob_0_teoritis, prob_gt_10_teoritis
```
